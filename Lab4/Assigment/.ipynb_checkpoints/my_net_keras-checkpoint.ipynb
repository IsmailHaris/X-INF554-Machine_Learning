# Network definition

```python
from time import clock
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from sklearn import multiclass, multioutput, neighbors, tree, base
```

```python
###############################################################
#
# Important notes: 
# - Do not change any of the existing functions or parameter names, 
#       except in __init__, you may add/change parameter names/defaults values.
# - In __init__ set default values to the best ones, e.g., learning_rate=0.1
# - Training epochs/iterations should not be a parameter to __init__,
#   To train/test your network, we will call fit(...) until time (2 mins) runs out.
#
###############################################################

class Network():

    def __init__(self, learning_rate=0.5, nb_layers=4):
        ''' Initialize the classifier with default (best) parameters '''

        # Parameters
        self.alpha       = learning_rate
        self.nb_layers   = nb_layers

        # Constants 
        self.n_labels      = 6
        self.start_epoch   = 0
        self.max_n_epochs  = 1000
        self.limit_reached = False

        if nb_layers == 2:
            network_layers = [keras.layers.Dense(128, activation=tf.nn.selu),
                              keras.layers.Dense(6, activation=tf.nn.sigmoid)]
        elif nb_layers == 3:
            network_layers = [keras.layers.Dense(128, activation=tf.nn.selu),
                              keras.layers.Dense(64, activation=tf.nn.selu),
                              keras.layers.Dense(6, activation=tf.nn.sigmoid)]
        elif nb_layers == 5:
            network_layers = [keras.layers.Dense(256, activation=tf.nn.selu),
                              keras.layers.Dense(128, activation=tf.nn.selu),
                              keras.layers.Dense(64, activation=tf.nn.selu),
                              keras.layers.Dense(32, activation=tf.nn.selu),
                              keras.layers.Dense(6, activation=tf.nn.sigmoid)]

        else: # Default, especially if nb_layers = 4
            network_layers = [keras.layers.Dense(128, activation=tf.nn.selu),
                              keras.layers.Dense(64, activation=tf.nn.selu),
                              keras.layers.Dense(32, activation=tf.nn.selu),
                              keras.layers.Dense(6, activation=tf.nn.sigmoid)]

        self.model = keras.Sequential(network_layers)

        self.model.compile(optimizer=tf.train.GradientDescentOptimizer(self.alpha), 
                           loss='binary_crossentropy',
                           metrics=['accuracy'])

    def fit(self,X,Y,warm_start=True,n_epochs=10):
        ''' Train the network, and if warm_start, then do not reinit. the network
            (if it has already been initialized)
        '''
        self.n_labels = Y.shape[1]

        if not warm_start: # Handle the warm start case
            self.start_epoch = 0
        
        if not self.limit_reached:
            self.model.fit(X, Y, epochs=n_epochs+self.start_epoch, initial_epoch=self.start_epoch)
            self.start_epoch += n_epochs
            self.limit_reached = self.start_epoch >= self.max_n_epochs

    def predict_proba(self,X):
        ''' Return a matrix P where P[i,j] = P(Y[i,j]=1), for all instances i, and labels j. '''
        return self.model.predict_proba(X)

    def predict(self,X):
        ''' Return a matrix of predictions for X '''
        return (self.predict_proba(X) >= 0.5).astype(int)
```

```python
###############################################################
# This is an example script, you may modify it as you wish
###############################################################

# Parameters
L = 6

# Load and parse the data (N instances, D features, L labels)
data = pd.read_csv('data/scene.csv') # Load data from CSV with Pandas
XY = data.values
N,DL = XY.shape
D = DL - L
Y = XY[:,0:L].astype(int)
X = XY[:,L:D+L]

# Split into train/test sets
n = int(N*6/10)
X_train = X[0:n]
Y_train = Y[0:n]
X_test = X[n:]
Y_test = Y[n:]

# Get class names
class_names = list(data.columns.values[:L])
```

# Main test

```python
t0 = clock()

# Test our classifier 
h = Network()

increment = 10
while (clock() - t0) < 120 and not h.limit_reached:
    h.fit(X_train,Y_train,n_epochs=increment)
```

```python
print("Trained %d epochs in %d seconds." % (h.start_epoch,int(clock() - t0)))
Y_pred = h.predict(X_test)
print(Y_test)
print(Y_pred)
loss = np.mean(Y_pred != Y_test)
print("Hamming loss: ", loss)
```

# First test

## For nb_layers and learning_rate

```python
def run_network(nb_layers=4, learning_rate=0.1, nb_epochs=600):
    
    print("\n\n########## Training NN with nb_layers={}, learning_rate={}, nb_epoch={} ##########".format(nb_layers,learning_rate,nb_epochs))

    t0 = clock()
    h = Network(learning_rate=learning_rate, nb_layers=nb_layers)

    h.fit(X_train,Y_train,n_epochs=nb_epochs, warm_start=False)

    exec_time = int(clock() - t0)

    Y_pred = h.predict(X_test)
    loss = np.mean(Y_pred != Y_test)

    return exec_time, loss
```

```python
nb_layers      = [2,3,4,5]
learning_rates = [0.0001, 0.001, 0.01, 0.1, 0.5, 1]

time_layers    = []
time_learning  = []

loss_layers    = []
loss_learning  = []
```

```python
for i in nb_layers:
    exec_time, loss = run_network(nb_layers=i)
    time_layers.append(exec_time)
    loss_layers.append(loss)
```

```python
for j in learning_rates:
    exec_time, loss = run_network(learning_rate=j)
    time_learning.append(exec_time)
    loss_learning.append(loss)
```

```python
plt.figure()
plt.plot(nb_layers, time_layers, 'r*-')
plt.xlabel('Number of layers')
plt.ylabel("Execution times")
plt.title("Test with learning_rate=0.1 & nb_epochs=600")
plt.show()

plt.figure()
plt.plot(nb_layers, loss_layers, 'b*-')
plt.xlabel('Number of layers')
plt.ylabel("Losses")
plt.title("Test with learning_rate=0.1 & nb_epochs=600")
plt.show()
```

```python
plt.figure()
plt.plot(learning_rates, time_learning, 'r*-')
plt.xlabel('Learning rates')
plt.ylabel("Execution times")
plt.title("Test with nb_layers=4 & nb_epochs=600")
plt.show()

plt.figure()
plt.plot(learning_rates, loss_learning, 'b*-')
plt.xlabel('Learning rates')
plt.ylabel("Losses")
plt.title("Test with nb_layers=4 & nb_epochs=600")
plt.show()
```

## For nb_epochs

```python
h = Network(learning_rate=0.5, nb_layers=4)
exec_time = 0

def run_network_for_epochs(nb_epochs):
    global exec_time

    t0 = clock()
    h.fit(X_train,Y_train,n_epochs=nb_epochs)
    exec_time += int(clock() - t0)

    Y_pred = h.predict(X_test)
    loss = np.mean(Y_pred != Y_test)

    return exec_time, loss
```

```python
nb_epochs       = [100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100]
nb_epochs_total = np.cumsum(nb_epochs)
time_epoch      = []
loss_epoch      = []
```

```python
for k in nb_epochs:
    exec_time, loss = run_network_for_epochs(nb_epochs=k)
    time_epoch.append(exec_time)
    loss_epoch.append(loss)
```

```python
plt.figure()
plt.plot(nb_epochs_total, time_epoch, 'r*-')
plt.xlabel('Number of epochs')
plt.ylabel("Execution times")
plt.title("Test with nb_layers=4 & learning_rate=0.1")
plt.show()

plt.figure()
plt.plot(nb_epochs_total, loss_epoch, 'b*-')
plt.xlabel('Number of epochs')
plt.ylabel("Losses")
plt.title("Test with nb_layers=4 & learning_rate=0.1")
plt.show()
```
